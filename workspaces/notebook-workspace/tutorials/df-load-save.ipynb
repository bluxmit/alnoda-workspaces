# Pandas dataframe
This simple example shows how to import and export dataframes from/to csv 

```python
import pandas as pd
```

```python
# Load csv file into Pandas data frame
df = pd.read_csv('cars.csv') 
```

```python
# Display pandas dataframe inline
df
```

```python
# add column
df['Liters_per_km'] = df['Miles_per_Gallon'] * 2.35215
```

```python
# Display pandas dataframe with new column
df
```

```python
# save transformed data frame to new .csv file
df.to_csv('cars-new.csv')
```
